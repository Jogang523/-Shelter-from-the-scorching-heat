```python
ptsd = df3['gu']
bar = 0.35

r1 = np.arange(len(ptsd))
r2 = r1 + bar

plt.figure(figsize=(18, 10))


plt.bar(r1, df3['합계'], color='y', width=bar,label='구별 인원 합계')


plt.bar(r2, df3['capacity'], color='b', width=bar, label='수용인원')


plt.title('구별 인원 합계 및 수용인원 비교')
# plt.xlabel('구')
# plt.ylabel('값')
plt.xticks([r + bar/2 for r in range(len(ptsd))], ptsd)


plt.legend()

plt.xticks(rotation = 45)
plt.tight_layout()
plt.show()

plt.savefig('C:/Users/M/siva/data_files년도별고령인구비율_추계.png', dpi = 300)
```
