```python
ㅁimport pandas as pd
gicho = pd.read_csv('C:/Users/M/Downloads/project/data/서울특별시 국민기초생활 수급자 동별 현황_20210731_v1.csv', encoding='cp949')
```

```python
gicho
```

```python
gicho_noin = gicho[gicho['연령구간'] == '65세이상']
gicho_noin2 = gicho_noin.groupby('시군구')['수급권자수'].sum().reset_index()

gicho_noin2
```

```python
gicho_noin2 = gicho_noin2.sort_values(by='수급권자수', ascending=False).reset_index(drop=True)

gicho_noin2
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import matplotlib.font_manager as fm

mpl.rcParams['axes.unicode_minus']= False
path = 'C:/Windows/Fonts/malgun.ttf'
font_name = fm.FontProperties(fname = path, size = 50).get_name()
plt.rc('font', family=font_name)

plt.figure(figsize=(10,8))
plt.bar(gicho_noin2['시군구'] , gicho_noin2['수급권자수'])
plt.xlabel('구별')
plt.xticks(rotation = 45)
plt.ylabel('수급권자 수')
plt.title('서울특별시 65세 이상 국민기초생활 수급자')
plt.savefig('C:/Users/M/Downloads/project/data/서울특별시 65세 이상 국민기초생활 수급자.png', dpi = 300)
```
