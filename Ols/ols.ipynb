```python
import pandas as pd
```

```python
df1=pd.read_csv('./data/등록인구(연령별_동별)_20240623184642.csv')
df1.drop('항목',axis=1,inplace=True)
df1.drop(0,inplace=True)
df1.columns=(['시군구','구별인구수'])
df1['구별인구수']=df1['구별인구수'].astype(int)
df1.head()
```

```python
df2=pd.read_csv('./data/Total_Population_by_District.csv')
df2.columns=(['시군구','고령자수'])
df2.head()
```

```python
df3=pd.read_csv('./data/국민건강보험공단_시군구별 온열질환자 수_20230630.csv',encoding='cp949')
df3=df3[(df3['진료년도']=='2022년')&(df3['시도']=='서울특별시')][['시군구','진료인원(명)']]
df3.columns=(['시군구','온열환자수'])
df3['온열환자수']=df3['온열환자수'].astype('int')
df3.head()
```

```python
df4=pd.read_csv('./data/무더위 쉼터 장소 정보.csv')
df4['구별']=df4.주소.apply(lambda x:x.split()[1])
df4.이용가능인원=df4.이용가능인원.apply(lambda x:x.replace('명',''))
df4.이용가능인원=df4.이용가능인원.astype('int')
df4=df4.groupby('구별')['이용가능인원'].sum().to_frame().reset_index()
df4.columns=(['시군구','수용인원'])
df4.head()
```

```python
df5=pd.read_csv('./data/구별 평균 최고 기온.csv',encoding='cp949')
df5.drop(columns = ['Unnamed: 0'], axis = 1, inplace = True)
df5.columns=(['시군구','최대기온'])
df5.head()
```

```python
df6=pd.read_csv('./data/구별 평균 기온.csv',encoding='cp949')
df6.drop(columns = ['Unnamed: 0'], axis = 1, inplace = True)
df6.columns=(['시군구','평균기온'])
df6.head()
```

```python
df7=pd.read_csv('./data/무더위 쉼터 장소 정보.csv')
df7['시군구']=df7.주소.apply(lambda x:x.split()[1])
df7=df7.groupby('시군구').count()['Unnamed: 0'].to_frame().reset_index()
df7.columns=(['시군구','무더위 쉼터 수'])
df7['무더위 쉼터 수']=df7['무더위 쉼터 수'].astype(int)
df7.head()
```

```python
df8=pd.read_csv('./data/result___.csv')
df8=df8.groupby('시군구')['수급권자수'].sum().to_frame().reset_index()
df8.head()
```

```python
df_merged = df1.merge(df2, how='inner')
df_merged = df_merged.merge(df3, how='inner')
df_merged = df_merged.merge(df4, how='inner')
df_merged = df_merged.merge(df5, how='inner')
df_merged = df_merged.merge(df6, how='inner')
df_merged = df_merged.merge(df7, how='inner')
df_merged = df_merged.merge(df8, how='inner')
```

```python
df_merged.info()
```

```python
df_merged.head()
```

```python
columns_to_scale = ['고령자수', '온열환자수', '수용인원','무더위 쉼터 수','수급권자수']
for column in columns_to_scale:
    df_merged[f'{column}_비율'] = df_merged[column] / df_merged['구별인구수']
```

```python
df_merged.to_csv('./data/회귀분석.csv')
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

result = ols('온열환자수_비율 ~ 고령자수_비율 + 수용인원_비율 + 수급권자수_비율 +평균기온 + 최대기온', data = df_merged).fit()
result.summary()
```
